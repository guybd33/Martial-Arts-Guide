```python
import cv2
import pandas as pd
import scipy.interpolate as interp
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
from tkinter import *



def comparison_algorithm():
    global tmp_score_acceleration
    df_map_teacher = pd.read_excel('Data_Analyzes_Teacher.xlsx')
    df_map_student = pd.read_excel('Data_Analyzes_Student.xlsx')
    df_angle_teacher = pd.read_excel('TeacherDataAngle.xlsx')
    df_angle_student = pd.read_excel('StudentDataAngle.xlsx')
    df_speed_teacher = pd.read_excel('TeacherDataSpeed.xlsx')
    df_speed_student = pd.read_excel('StudentDataSpeed.xlsx')
    df_acceleration_teacher = pd.read_excel('TeacherDataAcceleration.xlsx')
    df_acceleration_student = pd.read_excel('StudentDataAcceleration.xlsx')


    data_teacher_angle = []
    data_student_angle = []
    data_teacher_speed = []
    data_student_speed = []
    data_teacher_acceleration = []
    data_student_acceleration = []
    array_angles = []
    array_speed = []
    array_acceleration = []
    dived_moves_teacher = []
    dived_moves_student = []

    places_classifier = []
    fig = plt.figure()
    # ------------------------------sort the move to start and end move---------------
    j=0
    print(len(df_map_teacher['Unnamed: 0']))
    for j in range(0, len(df_map_teacher['Unnamed: 0']) - 1):
        if df_map_teacher.iloc[j + 1][1] > df_map_teacher.iloc[j][2] or df_map_teacher.iloc[j + 1][0] != \
                df_map_teacher.iloc[j][0]:
            dived_moves_teacher.append(j)
    dived_moves_teacher.append(j + 1)
    for j in range(0, len(df_map_student['Unnamed: 0']) - 1):
        if df_map_student.iloc[j + 1][1] > df_map_student.iloc[j][2] or df_map_student.iloc[j + 1][0] != \
                df_map_student.iloc[j][0]:
            dived_moves_student.append(j)
    dived_moves_student.append(j + 1)

    # ------------------------------sort the match of a move---------------
    counter_2 = 0
    counter_1 = 0
    r_of_student = 0
    k = 0
    match_move = 0
    print(dived_moves_teacher,dived_moves_student)
    for j in range(0, len(dived_moves_teacher)):
        for i in range(counter_2, dived_moves_teacher[j] + 1):
            flag = 0
            match_move = 0
            # p = len(df_map_student['Unnamed: 0'])
            if counter_1 > len(df_map_student['Unnamed: 0'])-1 :
                places_classifier.append([-1, -1])
            else:
                for k in range(counter_1, dived_moves_student[r_of_student] + 1):
                    if df_map_teacher.iloc[i][0] == df_map_student.iloc[k][0]:
                        match_move = 1
                        if df_map_teacher.iloc[i][3] == df_map_student.iloc[k][3]:
                            flag = 1
                            break

                if flag == 0 :
                    places_classifier.append([-1, -1])
                else:
                    match_move == 0
                    places_classifier.append([i, k])
                    counter_1 = k + 1
        if match_move == 1:
            places_classifier[len(places_classifier) - 1]=[i, k]
            # counter_1 = k + 1
        r_of_student += 1
        # if k == len()
        counter_2 = dived_moves_teacher[j] + 1
    print(places_classifier)

    # -------- map the data angles from the teacher to array---------------------------

    for j in range(0, len(df_map_teacher['Unnamed: 0'])):
        first = df_map_teacher.iloc[j, 1]
        finish = df_map_teacher.iloc[j, 2]
        data_first = 0
        data_last = 0
        for i in range(0, len(df_angle_teacher['Unnamed: 0'])):
            if df_angle_teacher.iloc[i, 20]-0.033 >= first and data_first == 0:
                data_first = i
            elif df_angle_teacher.iloc[i, 20]-0.033 >= finish and data_last == 0:
                data_last = i
                data_teacher_angle.append(
                    np.array(df_angle_teacher.iloc[data_first:data_last, df_map_teacher.iloc[j, 3]]))
                break


    # -------- map the data speed from the teacher to array---------------------------

    for j in range(0, len(df_map_teacher['Unnamed: 0'])):
        first = df_map_teacher.iloc[j, 1]
        finish = df_map_teacher.iloc[j, 2]
        data_first = 0
        data_last = 0
        for i in range(0, len(df_speed_teacher['Unnamed: 0'])):
            if df_speed_teacher.iloc[i, 20]-0.033 >= first and data_first == 0:
                data_first = i
            elif df_speed_teacher.iloc[i, 20]-0.033 >= finish and data_last == 0:
                data_last = i
                data_teacher_speed.append(
                    np.array(df_speed_teacher.iloc[data_first:data_last, df_map_teacher.iloc[j, 3]]))
                break

    # -------- map the data acceleration from the teacher to array---------------------------
    for j in range(0, len(df_map_teacher['Unnamed: 0'])):
        first = df_map_teacher.iloc[j, 1]
        finish = df_map_teacher.iloc[j, 2]
        data_first = 0
        data_last = 0
        for i in range(0, len(df_acceleration_teacher['Unnamed: 0'])):

            if df_acceleration_teacher.iloc[i, 20]-0.033 >= first and data_first == 0:
                data_first = i
            elif df_acceleration_teacher.iloc[i, 20]-0.033 >= finish and data_last == 0:
                data_last = i
                data_teacher_acceleration.append(
                    np.array(df_acceleration_teacher.iloc[data_first:data_last, df_map_teacher.iloc[j, 3]]))
                break
        if data_last==0:
            data_last = i
            data_teacher_acceleration.append(
                np.array(df_acceleration_teacher.iloc[data_first:data_last, df_map_teacher.iloc[j, 3]]))

    # -------- map the data angles from the student to array---------------------------

    for j in range(0, len(df_map_student['Unnamed: 0'])):
        first = df_map_student.iloc[j, 1]
        finish = df_map_student.iloc[j, 2]
        data_first = 0
        data_last = 0
        for i in range(0, len(df_angle_student['Unnamed: 0'])):
            if df_angle_student.iloc[i, 20] >= first and data_first == 0:
                data_first = i
            elif df_angle_student.iloc[i, 20] >= finish and data_last == 0:
                data_last = i
                data_student_angle.append(np.array(df_angle_student.iloc[data_first:data_last, df_map_student.iloc[j, 3]]))
                break
        if data_last==0:
            data_last = i
            data_student_angle.append(np.array(df_angle_student.iloc[data_first:data_last, df_map_student.iloc[j, 3]]))

    # -------- map the data speed from the student to array---------------------------


    for j in range(0, len(df_map_student['Unnamed: 0'])):
        first = df_map_student.iloc[j, 1]
        finish = df_map_student.iloc[j, 2]
        data_first = 0
        data_last = 0
        p = len(df_speed_student['Unnamed: 0'])
        for i in range(0, len(df_speed_student['Unnamed: 0'])):
            if df_speed_student.iloc[i, 20] >= first and data_first == 0:
                data_first = i
            elif df_speed_student.iloc[i, 20] >= finish and data_last == 0:
                data_last = i
                data_student_speed.append(
                    np.array(df_speed_student.iloc[data_first:data_last, df_map_student.iloc[j, 3]]))
                break
        if data_last==0:
            data_last = i
            data_student_speed.append(
                np.array(df_speed_student.iloc[data_first:data_last, df_map_student.iloc[j, 3]]))





    # -------- map the data acceleration from the student to array---------------------------

    for j in range(0, len(df_map_student['Unnamed: 0'])):
        first = df_map_student.iloc[j, 1]
        finish = df_map_student.iloc[j, 2]
        data_first = 0
        data_last = 0
        for i in range(0, len(df_acceleration_student['Unnamed: 0'])):
            if df_acceleration_student.iloc[i, 20]-0.033 >= first and data_first == 0:
                data_first = i
            elif df_acceleration_student.iloc[i, 20]-0.033 >= finish and data_last == 0:
                data_last = i
                data_student_acceleration.append(
                    np.array(df_acceleration_student.iloc[data_first:data_last, df_map_student.iloc[j, 3]]))
                break
        if data_last==0:
            data_last = i
            data_student_acceleration.append(
                np.array(df_acceleration_student.iloc[data_first:data_last, df_map_student.iloc[j, 3]]))


    # -------- interpolat and crosscor  data student to teacher to array---------------------------

    for j in range(0, len(places_classifier)):
        if places_classifier[j] == [-1, -1]:
            array_angles.append(0)
        else:

            arr1_interp = interp.interp1d(np.arange(data_student_angle[places_classifier[j][1]].size), data_student_angle[places_classifier[j][1]])
            arr1_compress = arr1_interp(
                np.linspace(0, data_student_angle[places_classifier[j][1]].size - 1, data_teacher_angle[places_classifier[j][0]].size))
            x = plt.xcorr(arr1_compress, data_teacher_angle[places_classifier[j][0]], maxlags=len(arr1_compress) - 1)
            array_angles.append(x[1][len(arr1_compress) - 1])



    # -------- interpolat and crosscor  speed student to teacher to array---------------------------


    for j in range(0, len(places_classifier)):
        if places_classifier[j] == [-1, -1]:
            array_speed.append(0)
        else:

            arr1_interp = interp.interp1d(np.arange(data_student_speed[places_classifier[j][1]].size), data_student_speed[places_classifier[j][1]])
            arr1_compress = arr1_interp(
                np.linspace(0, data_student_speed[places_classifier[j][1]].size - 1, data_teacher_speed[places_classifier[j][0]].size))
            x = plt.xcorr(arr1_compress, data_teacher_speed[places_classifier[j][0]], maxlags=len(arr1_compress) - 1)
            array_speed.append(x[1][len(arr1_compress) - 1])


    # -------- interpolat and crosscor  acceleration student to teacher to array---------------------------
    for j in range(0, len(places_classifier)):
        if places_classifier[j] == [-1, -1]:
            array_acceleration.append(0)
        else:
            arr1_interp = interp.interp1d(np.arange(data_student_acceleration[places_classifier[j][1]].size), data_student_acceleration[places_classifier[j][1]])
            arr1_compress = arr1_interp(
                np.linspace(0, data_student_acceleration[places_classifier[j][1]].size - 1, data_teacher_acceleration[places_classifier[j][0]].size))
            x = plt.xcorr(arr1_compress, data_teacher_acceleration[places_classifier[j][0]], maxlags=len(arr1_compress) - 1)
            array_acceleration.append(x[1][len(arr1_compress) - 1])


    plt.close()


    score_acceleration = 0
    score_angles = 0
    score_speed = 0
    counter = 0
    for j in range(0, len(dived_moves_teacher)):
        length = 0
        tmp_score_acceleration = 0
        tmp_score_angle = 0
        tmp_score_speed = 0
        for i in range(counter, dived_moves_teacher[j]+1):
            if array_angles[i] != 0:
                tmp_score_acceleration = abs(array_acceleration[i]) + tmp_score_acceleration
                # tmp_score_time = time_teacher[i] + tmp_score_time
                tmp_score_angle = array_angles[i] + tmp_score_angle
                tmp_score_speed = array_speed[i] + tmp_score_speed
                length += 1
        if length != 0:
            score_acceleration = (tmp_score_acceleration/length) + score_acceleration
            score_speed = (tmp_score_speed / length) + score_speed
            score_angles = (tmp_score_angle / length) + score_angles
        counter = i+1




    score = ((score_speed * 40) + (score_angles * 40) + (score_acceleration*20)) / len(dived_moves_teacher)

    comparison_gui = tk.Tk()
    comparison_gui.title("Scores")  # window name
    comparison_gui.config(background="gray")  # background color
    comparison_gui.maxsize(700, 700)


    ws = comparison_gui.winfo_screenwidth()  # width of the screen
    hs = comparison_gui.winfo_screenheight()  # height of the screen

    w = 800
    h = 650
    x = (ws/ 4) - (w / 2)
    y = (hs / 1.3) - (h / 2)
    comparison_gui.geometry('%dx%d+%d+%d' % (w, h, x, y))

    main_name = Label(comparison_gui, text="Scores", font=('Arial', 20, 'bold'), fg='green', bg='black',
                      relief=RAISED, bd=10)  # create label
    main_name.place(x=200, y=1)  # put the label in gui

    speed_score = Label(comparison_gui, text="speed score = " + str(round((score_speed / len(dived_moves_teacher) )* 40, 3)) + "/40",
                        font=('Arial', 20, 'bold'), fg='green', bg='black',
                        relief=RAISED, bd=10)  # create label
    speed_score.place(x=30, y=60)  # put the label in gui

    angle_score = Label(comparison_gui, text="angle score = " + str(round((score_angles/ len(dived_moves_teacher) ) * 40, 3)) + "/40",
                        font=('Arial', 20, 'bold'),
                        fg='green', bg='black',
                        relief=RAISED, bd=10)  # create label
    angle_score.place(x=30, y=120)  # put the label in gui

    acceleration_score = Label(comparison_gui,
                               text="acceleration score = " + str(round((score_acceleration/ len(dived_moves_teacher) ) * 20, 3)) + "/20",
                               font=('Arial', 20, 'bold'),
                               fg='green', bg='black',
                               relief=RAISED, bd=10)  # create label
    acceleration_score.place(x=30, y=180)  # put the label in gui

    total_score = Label(comparison_gui, text="total  score = " + str(round(score, 3)) + "%",
                        font=('Arial', 20, 'bold'),
                        fg='green', bg='black',
                        relief=RAISED, bd=10)  # create label
    total_score.place(x=30, y=240)  # put the label in gui

    exit_button = Button(comparison_gui, text='Close', width=10, height=2, borderwidth=3,
                         command=comparison_gui.destroy)
    exit_button.place(x=465, y=350)

    plot_button = Button(master=comparison_gui,
                         command= lambda: plot(df_map_teacher, df_map_student, data_teacher_angle, data_student_angle, data_teacher_speed,
                                               data_student_speed, data_teacher_acceleration, data_student_acceleration, places_classifier),
                         height=2,
                         width=10,
                         text="Plot")

    plot_button.place(x=200, y=350)

    comparison_gui.mainloop()


def plot(df_map_teacher,df_map_student,data_teacher_angle,data_student_angle,data_teacher_speed,
        data_student_speed,data_teacher_acceleration,data_student_acceleration,a):

    color_arr = ['r', 'g', 'b', 'c', 'm', 'y', 'k']

    fig, axs = plt.subplots(len(a))
    fig.suptitle('Angles')
    p = 0 ##index for colors
    for i in range (0,len(a)):
        if a[i] == [-1, -1]:
            axs[i].text(0.6, 0.7, "could not find a match", size=20,
                        ha="center", va="center",
                        bbox=dict(boxstyle="round",
                                  ec=(1., 0.5, 0.5),
                                  fc=(1., 0.8, 0.8),
                                  )
                        )
        else:
            t = np.linspace(df_map_teacher.iloc[a[i][0], 1],df_map_teacher.iloc[a[i][0], 2],len(data_teacher_angle[a[i][0]]))
            axs[i].plot(t,data_teacher_angle[a[i][0]],c = color_arr[p],label='teacher, '+str(i+1)+' comparison, ' + df_map_teacher.iloc[a[i][0], 0]+
                                                                                        ' ,angle num: '+ str(df_map_teacher.iloc[a[i][0], 3]))
            t = np.linspace(df_map_teacher.iloc[a[i][0], 1],df_map_teacher.iloc[a[i][0], 2],len(data_student_angle[a[i][1]]))
            axs[i].plot(t,data_student_angle[a[i][1]],linestyle = 'dotted', c=color_arr[p],label='student, '+str(i+1)+' comparison, ' + df_map_student.iloc[a[i][1], 0]+
                                                                                       ' ,angle num: '+ str(df_map_student.iloc[a[i][1], 3]))
            axs[i].legend()
            axs[i].set_xlabel('Time')
            axs[i].set_ylabel('[Radians]')
        p = p +1
        if p==7:
            p=0

    fig, axs = plt.subplots(len(a))
    fig.suptitle('Speeds')
    p = 0  ##index for colors
    for i in range (0,len(a)):
        if a[i] == [-1, -1]:
            axs[i].text(0.6, 0.7, "could not find a match", size=20,
                        ha="center", va="center",
                        bbox=dict(boxstyle="round",
                                  ec=(1., 0.5, 0.5),
                                  fc=(1., 0.8, 0.8),
                                  )
                        )
        else:
            t = np.linspace(df_map_teacher.iloc[a[i][0], 1],df_map_teacher.iloc[a[i][0], 2],len(data_teacher_speed[a[i][0]]))
            axs[i].plot(t,data_teacher_speed[a[i][0]],c = color_arr[p],label='teacher, '+str(i+1)+' comparison, ' + df_map_teacher.iloc[a[i][0], 0]+
                                                                                        ' ,angle num: '+ str(df_map_teacher.iloc[a[i][0], 3]))
            t = np.linspace(df_map_teacher.iloc[a[i][0], 1],df_map_teacher.iloc[a[i][0], 2],len(data_student_speed[a[i][1]]))
            axs[i].plot(t,data_student_speed[a[i][1]],linestyle = 'dotted', c=color_arr[p],label='student, '+str(i+1)+' comparison, ' + df_map_student.iloc[a[i][1], 0]+
                                                                                                   ' ,angle num: '+ str(df_map_student.iloc[a[i][1], 3]))
            axs[i].set_xlabel('Time')
            axs[i].set_ylabel('[m/s]')
            axs[i].legend()
        p = p + 1
        if p == 7:
            p = 0

    fig, axs = plt.subplots(len(a))
    fig.suptitle('Accelerations')
    p = 0  ##index for colors
    for i in range (0,len(a)):
        if a[i] == [-1, -1]:
            axs[i].text(0.6, 0.7, "could not find a match", size=20,
                        ha="center", va="center",
                        bbox=dict(boxstyle="round",
                                  ec=(1., 0.5, 0.5),
                                  fc=(1., 0.8, 0.8),
                                  )
                        )
        else:
            t = np.linspace(df_map_teacher.iloc[a[i][0], 1],df_map_teacher.iloc[a[i][0], 2],len(data_teacher_acceleration[a[i][0]]))
            axs[i].plot(t,data_teacher_acceleration[a[i][0]],c = color_arr[p],label='teacher, '+str(i+1)+' comparison, ' + df_map_teacher.iloc[a[i][0], 0]+
                                                                                                     ' ,angle num: '+ str(df_map_teacher.iloc[a[i][0], 3]))
            t = np.linspace(df_map_teacher.iloc[a[i][0], 1],df_map_teacher.iloc[a[i][0], 2], len(data_student_acceleration[a[i][1]]))
            axs[i].plot(t,data_student_acceleration[a[i][1]],linestyle = 'dotted', c=color_arr[p],label='student, '+str(i+1)+' comparison, ' + df_map_student.iloc[a[i][1], 0]+
                                                                                              ' ,angle num: '+ str(df_map_student.iloc[a[i][1], 3]))
            axs[i].set_xlabel('Time')
            axs[i].set_ylabel('[m/s^2]')
            axs[i].legend()
        p = p + 1
        if p == 7:
            p = 0

    plt.show()
```
